```python
#Toy Store Inventory
Inventory = {101:{"Name":"Teddy Bear", "price":600,"stock":75},102:{"name":"Rubik's Cube","price":200,"stock":40}, 103:{"name":"RC Car","price":1200,"stock":65}}
#A loop to bring up the menu again after the operation is performed till the user decides to end the program.
end=False
while(end==False):
    #The choice to select the operation to be performed
    print("Press 1 to Add a new Toy.")
    print("Press 2 to Check Toy Availablity.")
    print("Press 3 to Retrieve a Toy's Price.")
    print("Press 4 to Update the stock of a Toy.")
    print("Press 5 to Remove a Toy.")
    print("Press 6 to List all Toys with details")
    print("Press 7 to Clear the inventory.")
    print("Press 0 to Exit")
    choice=int(input("Select the operation to be performed:"))
    print("\n")
    if(choice==0):
        end=True
        print("The program will now terminate.")
    elif(choice==1):
        toy_id=int(input("Enter new Toy ID:"))
        name=input("Enter Toy's Name:")
        price=float(input("Enter Toy's Price:"))
        stock = int(input("Enter stock quantity:"))
        Inventory[toy_id]={"name":name,"price":price,"stock":stock}
        print("\nUpdated Inventory:")
        print(Inventory)
        print("\n")
    elif(choice==2):
        toy_id=int(input("Enter Toy ID to check:"))
        if toy_id in Inventory:
            print("Toy Available:",Inventory[toy_id]["name"])
        else:
            print("Toy not found.")
        print("\n")
    elif(choice==3):
        toy_id=int(input("Enter Toy ID:"))
        if toy_id in Inventory:
            print("Price:",Inventory[toy_id]["price"])
        else:
            print("Toy ID not found.")
        print("\n")
    elif(choice==4):
        toy_id=int(input("Enter Toy ID:"))
        if toy_id in Inventory:
            print("1. Increase Stock.")
            print("2. Decrease Stock.")
            opt = int(input("Enter choice"))
            qty = int(input("Enter Quantity:"))
            if(opt==1):
                Inventory[toy_id]["stock"]+=qty
                print("Updated Stock:",Inventory[toy_id]["stock"])
            elif(opt==2):
                Inventory[toy_id]["stock"]-=qty
                print("Updated Stock:",Inventory[toy_id]["stock"])
        else:
            print("Toy not found.")
        print("\n")
    elif(choice==5):
        toy_id=int(input("Enter Toy ID to remove:"))
        if Inventory.pop(toy_id,None):
            print("Toy removed successfully.")
        else:
            print("Toy ID not found.")
        print("\nUpdated Inventory:")
        print(Inventory)
        print("\n")
    elif(choice==6):
        print("\nAll Toys in inventory:")
        for t_id,details in Inventory.items():
            print(t_id,":",details)
        print("\n")
    elif(choice==7):
        Inventory.clear()
        print("Inventory Cleared.")
        print(Inventory)
        print("\n")
    
```
